```python
from selenium import webdriver
import time
from time import sleep

driver = webdriver.Chrome('chromedriver')
driver.implicitly_wait(2)

#멜론 차트파인더 접속
driver.get('https://www.melon.com/')
driver.find_element_by_class_name('nth1').click()
driver.implicitly_wait(2)
driver.find_element_by_class_name('btn_chart_f').click()

#주간차트 클릭

if driver.find_element_by_class_name('tab01'):
    driver.find_element_by_class_name('tab01').click()
driver.implicitly_wait(2)
sleep(2)
#연도별 //2020년대 ...li[1], 2010년대 ...li[2]
year_big = driver.find_element_by_xpath('//*[@id="d_chart_search"]/div/div/div[1]/div[1]/ul')
year_big_li = year_big.find_elements_by_tag_name('li')
year_big_len = len(year_big_li)
driver.implicitly_wait(2)
for i1 in range(2,year_big_len+1):
# for i1 in range(1, year_big_len+1) :
    try:    
        driver.find_element_by_class_name('tab01').click()
    except :
        pass
    driver.implicitly_wait(2)
    driver.find_element_by_xpath('//*[@id="d_chart_search"]/div/div/div[1]/div[1]/ul/li[%d]' % i1).click()
    #연도 //2021년 ...li[1], 2020년 ...li[2]
    year_small = driver.find_element_by_xpath('//*[@id="d_chart_search"]/div/div/div[2]/div[1]/ul')
    year_small_li = year_small.find_elements_by_tag_name('li')
    year_small_len =len(year_small_li)
    driver.implicitly_wait(2)
    sleep(1)
    for i2 in range(2, 3):
#     for i2 in range(1, year_small_len+1):
        if driver.find_element_by_class_name('tab01'):
            driver.find_element_by_class_name('tab01').click()
        driver.implicitly_wait(2)
        sleep(1)
        driver.find_element_by_xpath('//*[@id="d_chart_search"]/div/div/div[1]/div[1]/ul/li[%d]' % i1).click()
        driver.find_element_by_xpath('//*[@id="d_chart_search"]/div/div/div[2]/div[1]/ul/li[%d]' % i2).click()
        month = driver.find_element_by_xpath('//*[@id="d_chart_search"]/div/div/div[3]/div[1]/ul')
        month_li = month.find_elements_by_tag_name('li')
        month_len = len(month_li)
        
        for i3 in range(11, month_len+1):
#         for i3 in range(1, month_len+1):
            if driver.find_element_by_class_name('tab01'):
                driver.find_element_by_class_name('tab01').click()
            driver.implicitly_wait(2)
            sleep(0.3)
            driver.find_element_by_xpath('//*[@id="d_chart_search"]/div/div/div[1]/div[1]/ul/li[%d]' % i1).click()
            driver.find_element_by_xpath('//*[@id="d_chart_search"]/div/div/div[2]/div[1]/ul/li[%d]' % i2).click()
            driver.find_element_by_xpath('//*[@id="d_chart_search"]/div/div/div[3]/div[1]/ul/li[%d]' % i3).click()
            week = driver.find_element_by_xpath('//*[@id="d_chart_search"]/div/div/div[4]/div[1]/ul')
            week_li = week.find_elements_by_tag_name('li')
            week_len = len(week_li)
            driver.implicitly_wait(2)
            sleep(0.3)
            fw = open(str(2030-i1*10+year_small_len-i2)+'_'+str(i3)+'.txt','w')
            for i4 in range(1, week_len+1):
                for i5 in range(1, 101):
                    if driver.find_element_by_class_name('tab01'):
                        driver.find_element_by_class_name('tab01').click()
                    driver.implicitly_wait(2)
                    sleep(0.3)
                    
                    driver.find_element_by_xpath('//*[@id="d_chart_search"]/div/div/div[1]/div[1]/ul/li[%d]' % i1).click()
                    driver.implicitly_wait(2)
                    sleep(0.3)
                    driver.find_element_by_xpath('//*[@id="d_chart_search"]/div/div/div[2]/div[1]/ul/li[%d]' % i2).click()
                    driver.implicitly_wait(2)
                    sleep(0.3)
                    driver.find_element_by_xpath('//*[@id="d_chart_search"]/div/div/div[3]/div[1]/ul/li[%d]' % i3).click()
                    driver.implicitly_wait(2)
                    sleep(0.3)
                    driver.find_element_by_xpath('//*[@id="d_chart_search"]/div/div/div[4]/div[1]/ul/li[%d]' % i4).click()
                    driver.implicitly_wait(2)
                    sleep(0.3)
                    driver.find_element_by_xpath('//*[@id="d_chart_search"]/div/div/div[5]/div[1]/ul/li[1]').click()
                    driver.implicitly_wait(2)
                    sleep(0.3)
                    print(driver.find_element_by_xpath('//*[@id="d_chart_search"]/div/div/div[2]/div[1]/ul/li[%d]' % i2).text)
                    print(driver.find_element_by_xpath('//*[@id="d_chart_search"]/div/div/div[3]/div[1]/ul/li[%d]' % i3).text)
                    print(driver.find_element_by_xpath('//*[@id="d_chart_search"]/div/div/div[4]/div[1]/ul/li[%d]' % i4).text)
                    print(i5)
                    print('-----------------------------------------------')
                    driver.find_element_by_class_name('btn_b26').click()
                    driver.implicitly_wait(2)
                    sleep(0.3)
                    if i5 > 50 :
                        driver.find_element_by_xpath('//*[@id="frm"]/div[2]/span/a').click()
                        driver.implicitly_wait(2)
                        sleep(0.3)
                    driver.find_element_by_xpath('/html/body/div/div[3]/div/div/div/div[1]/div[2]/form/div[1]/table/tbody/tr[%d]/td[4]/div/a' % i5).click()
                    sleep(0.3)
                    try: 
                        r = driver.find_element_by_xpath('//*[@id="d_video_summary"]')
                        try : fw.write(r.text)
                        except : pass
                    except: pass
                    driver.back()
                    sleep(0.5)
            fw.close()
                    
                    
                

                
```
